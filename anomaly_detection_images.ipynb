```python
# Show a few images from the dataset

def load_images_from_folder(folder):
    
    collect_images = []
    for filename in os.listdir(folder):
        read_image = cv2.imread(os.path.join(folder,filename))
        if read_image is not None:
            collect_images.append(read_image)
        if len(collect_images)>7:
            break

    for i in range(len(collect_images)):
        plt.imshow(collect_images[i])
        plt.show()
        i+=1
```

```python
# Plot images with cracks (Positive Samples)

load_images_from_folder(base_path + "Positive")
```

```python
# Plot images without cracks (Negative Samples)

load_images_from_folder(base_path + "Negative")
```
